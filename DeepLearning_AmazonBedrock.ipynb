```python
# the template from tutorial

template = """
I need to summarize a presentation. The transcript of the 
presentation is between the <data> XML like tags.

<data>
{transcript}
</data>

The output should be in JSON format and contain a sentiment analysis of the presentation and
its summarized version. The summary should not be longer than 200 words.

Example output:
{{
    "sentiment": <sentiment>,
    "summary": <summary>
}}

Write the JSON output and nothing more.

Here is the JSON output:
"""

json.dumps(template)
```

```python
templates_obj = {
    "templates": [
        {"name": "summary_template_tutorial", "content": template}
    ]
}
with open('./data/sound/summary-templates.json', 'w') as f:
    json.dump(templates_obj, f)
```

```python
with open('./data/sound/summary-templates.json', 'r') as f:
    summary_templates = json.load(f)
```

```python
prompt = None
tmpl = next((el for el in summary_templates['templates'] if el['name'] == 'summary_template_tutorial'), None)
if tmpl is not None:
    prompt = tmpl['content'].format(transcript=transcript)
```

```python
prompt
```

```python
kwargs = {
    "modelId": "amazon.titan-text-express-v1",
    "contentType": "application/json",
    "accept": "*/*",
    "body": json.dumps(
        {
            "inputText": prompt,
            "textGenerationConfig": {
                "maxTokenCount": 256,
                "temperature": 0.1,
                "topP": 0.8
            }
        }
    )
}
response = bedrock_runtime.invoke_model(**kwargs)
```

```python
response_body = json.loads(response['body'].read().decode('utf-8'))
```
